## TABELA - CAMPEONATO BRASILEIRO

```python
import pandas as pd
import requests
import re
import plotly.graph_objects as go

# Extração inicial
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
url = r'https://www.espn.com.br/futebol/classificacao/_/liga/bra.1'
request = requests.get(url, headers=header)

dfs = pd.read_html(request.text)

# Extrair os DataFrames de interesse
Times = dfs[0].copy()
Pontos = dfs[1].copy()

# Sem iniciais dos clubes
Times.iloc[:, 0] = Times.iloc[:, 0].apply(lambda x: re.search(r'[A-Z][a-z].*', str(x)).group() if re.search(r'[A-Z][a-z]', str(x)) else x)

# Remover números da primeira coluna de "Times"
Times.iloc[:, 0] = Times.iloc[:, 0].apply(lambda x: re.sub(r'\d+', '', str(x)))

# Montando a tabela
tabela = Times.merge(Pontos, how='inner', left_index=True, right_index=True)
tabela.reset_index(drop=True, inplace=True)
tabela.index = Times.index + 1
tabela.rename(columns={'2024': 'Times'})
```
